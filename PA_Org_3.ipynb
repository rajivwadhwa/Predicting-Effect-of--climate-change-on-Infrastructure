```python
def grab_decade(start_yr, y_c_data, interval=10):
    '''Return years and counts for only a specific interval length.'''
    end_yr = int(start_yr) + interval - 1
    years = y_c_data[(y_c_data['YEAR'] <= end_yr) & (y_c_data['YEAR'] >= start_yr)]
    return years
```
